```python
from fredata import series,window_equalize
import matplotlib.pyplot as plt
import numpy as np
import urllib
%matplotlib inline
```

```python
# Verify that initialization works
gdp = series('GDPC1')
cpi = series('CPIAUCSL')

gdp.apc()
cpi.apc()

window_equalize([gdp,cpi])

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot_date(gdp.datenumbers,gdp.data,'b-',lw=2)
plt.plot_date(cpi.datenumbers,cpi.data,'r-',lw=2)
plt.legend(['gdp','cpi'],loc='upper right')
ax.set_ylabel('percent')
ax.grid(True)
```
